# Json 형식의 파일 크롤링

```python
import json, re
from urllib.request import urlopen
from html import unescape
```

#### 웹 페이지 읽어오기

```python
request = urlopen('http://www.hanbit.co.kr/store/books/full_book_list.html')
request.info().get_content_charset('utf-8')
html = request.read().decode(encoding='utf-8')
html[0:1000]
```

- json.dumps() : 데이터를 json 형식으로 인코딩 (문자열)
- json.dump() : 데이터를 json 형식으로 인코딩하여 파일에 출력 (파일)
- ensure_ascii = False : \xxxx 형태로 이스케이프하지 않고 출력
- [{ key:value}, {key:value}] indent=size

- 정규 표현식
- . : 모든 문자 한글자
- \*  : 0번 이상 반복
- ? : 0번 또는 1번
- < a href='( .*?)' > : href에 문자가 없어도 됨 & 하나의 문자가 반복됨
- < td class = 'left' > < a *? < /td > : a로 시작하는 문자가 반복되거나 없음


#### json 파일 생성

```python
with open("./data/booklist_json.json", mode="w", encoding="utf-8") as file:
    data = []
    for partial_html in re.findall(r'<td class="left"><a.*?</td>', html):
        # group() : 매칭되는 태그만 가져옴
        # group(1) : re함수에서 괄호로 묶은 것 안에 있는 요소들만 가져옴
        search = re.search(r'<a href="(.*?)">', partial_html).group(1)
        url = 'http://www.hanbit.co.kr'+search
        title = re.sub(r'<.*?>','', partial_html)
        data.append({'bookName' : title, 'link' : url})
        print(json.dumps(data, ensure_ascii = False, indent=2))
    json.dump(data, file, ensure_ascii = False, indent=2)
```

## Selenium

#### selenium 특정 요소를 찾는 함수 
- find_element_() : 단일 태그를 찾음 (= bs의 find 함수)
- find_elements_() : 다중 태그를 찾음 (= bs의 find_all 함수)
- find_element_by_id() : id의 속성을 찾음
- find_element_by_name() : 태그 이름으로 찾음
- find_element_by_css_selector() : css(class)로 찾음

- web의 element 객체를 리턴 : text

## 네이버 지도에서 '치킨'을 검색한 뒤 가게명, 주소, 번호 크롤링

```python
from selenium import webdriver
import time
path = "C:/selenium/chromedriver.exe" 
driver = webdriver.Chrome(path)
driver.get('https://v4.map.naver.com/')
```

```python
driver.find_element_by_css_selector('#dday_popup > div.popup_content.popup_link > button').click()
search_box = driver.find_element_by_css_selector('input#search-input')
search_box.send_keys('치킨')
search_button = driver.find_element_by_css_selector('button.spm')
search_button.click()
time.sleep(1)
```

```python
lists = driver.find_elements_by_css_selector('div.lsnx')
names = driver.find_elements_by_css_selector('dt > a')
addrs = driver.find_elements_by_css_selector('dd.addr')
tels = driver.find_elements_by_css_selector('dd.tel')
store_name = []
store_addr = []
store_tel = []
for name in names:
    store_name.append(name.text)
for addr in addrs:
    store_addr.append(addr.text)
for tel in tels:
    store_tel.append(tel.text)
```

```python
stores = driver.find_elements_by_css_selector('div.lsnx')
num = 0
for store in stores:
    num += 1
    if num == 7:
        name = store.find_element_by_css_selector('dt > a').text
        addr = store.find_element_by_css_selector('dd.addr').text
        print(name, addr, ' 번호 정보 없음 ')
    else:
        name = store.find_element_by_css_selector('dt > a').text
        addr = store.find_element_by_css_selector('dd.addr').text
        tel = store.find_element_by_css_selector('dd.tel').text
        print(name, addr, tel)
```
